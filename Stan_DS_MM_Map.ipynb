# Stan's Miracle Message Map Data Clean_up.

```python
# Reading in as DataFrame from Matt Kirby's dataset csv.

import pandas as pd

mm = pd.read_csv('https://raw.githubusercontent.com/mkirby1995/miracle_messages/master/Master%20MM%20Volunteers%20-%20MM%20Volunteers.csv')
pd.set_option('display.max_columns', 24)
mm.head()
```

```python
mm = mm.drop(columns=['Donations', 'Donation Date', 'Donation Source', 'Source', 'Company', 'Emails sent', 'Opened', 'Clicked', 'Birthday', 'Name', 'Merge status'])

mm.head()
```

```python
mm.shape
```

```python
# !pip install validate_email
```

```python
from validate_email import validate_email

mm['Valid Email?'] = mm['Email address'].apply(lambda x:validate_email(x))

print('All rows with valid email addresses:', mm['Valid Email?'].value_counts())
```

```python
mm['Has Comment'] = mm['Comment'].notnull()

mm.head()
```

```python
mm["Comment Contains: 'interested'"] = mm['Comment'].str.contains('interested', case=False, na=False) == True

mm["Comment Contains: 'interested'"].value_counts()[1:2]
```
